채팅방에 들어오고 나가거나, 닉네임을 변경한 기록이 담긴 문자열 배열 record가 매개변수로 주어질 때, 

모든 기록이 처리된 후, 최종적으로 방을 개설한 사람이 보게 되는 메시지를 문자열 배열 형태로 return 하도록 solution 함수를 완성하라.

```python
record = ["Enter uid1234 Muzi", "Enter uid4567 Prodo","Leave uid1234","Enter uid1234 Prodo","Change uid4567 Ryan"]


def solution(record):
    answer = []
    nickname_dic = {}
    action_dic = {"Enter":"들어왔습니다."}

    for log in record:

        ain = log.split()

        if len(ain) == 3:

            action, id, nickname = ain

            if action == "Enter" or action == "Change":

                nickname_dic[id] = nickname
        else:

            continue

        
    for log in record:

        if "Change" not in log:

            log_split = log.split()

            if len(log_split) == 3:

                action, id, nickname = log_split

                answer.append("{}님이 {}".format(nickname_dic[id], action_dic[action]))
            
            else:

                action, id = log_split
                answer.append("{}님이 나갔습니다.".format(nickname_dic[id]))

    return answer

solution(record)
```
