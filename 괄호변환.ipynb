```python
def divide(p): # 균형 잡힌 괄호 문자열 만들기
    
    left, right = 0, 0
    
    for i in range(len(p)):
        
        if p[i] == '(':
            left += 1
        else:
            right += 1
            
        if left == right: # (과 )의 수가 같다면 균형 잡힌 괄호 문자열
            
            return p[:i+1], p[i+1:]
        
def correct(p): # 올바른 괄호 문자열인지 확인
    
    stack = []
    
    for s in p:
        
        if s == '(':
            stack.append(s)
        else:
            if not stack:
                return False
            else:
                stack.pop()
                
    return True


def solution(p):
    
    answer = ""
    
    # STEP 1.
    if not p: # 빈 문자열이면 그대로 반환
        return answer
    
    # STEP 2: 균형잡힌 괄호 문자열로 분리
    u,v = divide(p)
    
    # STEP 3
    if correct(u): # u가 올바른 괄호 문자열이면
        # 3-1. v에 대해 동일한 과정 수행
        return u + solution(v)
    
    # STEP 4
    else:
        # 4-1, 4-2, 4-3
        answer += "("
        answer += solution(v)
        answer += ")"
        
        # 4-4
        for i in u[1:len(u)-1]:
            if i == "(":
                answer += ")"
            else:
                answer += "("
    
    return answer
```
